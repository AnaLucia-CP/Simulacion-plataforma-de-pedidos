```python
import simpy
import random
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
from geopy.distance import great_circle
import pandas as pd
import os
import openpyxl
from openpyxl.drawing.image import Image

class RechazoDomiciliario:
    def __init__(self):
        self.probabilidad = 0
        
    def establecer_probabilidad(self, nueva_probabilidad):
        self.probabilidad = nueva_probabilidad

    def rechazar(self):
        return random.random() <= self.probabilidad

class Plataforma:
    def __init__(self, env, domiciliarios, tiendas, platform_time):
        self.env = env
        self.domiciliarios = domiciliarios
        self.tiendas = tiendas
        self.platform_time = platform_time
        self.tiempo_disponibilidad = {} # Aquí almacenaremos los tiempos de disponibilidad de los domiciliarios
        self.rechazo_domiciliario = RechazoDomiciliario()
        self.pedidos_rechazados = 0
        self.ganancia_total_domiciliarios = 0
        self.ganancia_total_tiendas = 0
        self.ganancia_total_plataforma = 0
        self.ventas_totales = 0
        self.ordenes_perdidas = []  # Lista para almacenar las órdenes perdidas
        self.rechazos_por_orden = {}  # Diccionario para llevar el conteo de rechazos por orden
        self.ganancias_perdidas_tiendas = 0
        self.ganancias_perdidas_plataforma = 0
        self.ganancias_perdidas_domiciliarios = 0
        self.pago_tienda_a_plataforma = 0
        self.registro_domiciliarios = {}


    def obtener_tienda_por_id(self, store_id):
        tienda = self.tiendas.loc[self.tiendas['store_id'] == store_id]
        if not tienda.empty:
            return tienda.iloc[0]
        return None

    def obtener_domiciliarios_disponibles(self, orden, capacidad_domiciliario):
        domiciliarios_disponibles = []

        tiempo_actual = orden.placement_time

        for index, domiciliario in self.domiciliarios.iterrows():
            domiciliario_id = domiciliario['courier_id']
            tiempo_disponibilidad = self.tiempo_disponibilidad.get(domiciliario_id, tiempo_actual)

            dist = great_circle((orden.drop_off_lat, orden.drop_off_lng), (domiciliario['on_lat'], domiciliario['on_lng'])).kilometers

            if tiempo_disponibilidad <= tiempo_actual and dist <= capacidad_domiciliario:
                domiciliarios_disponibles.append(domiciliario)

        return domiciliarios_disponibles

    def procesar_orden(self, orden, lista_salida, lista_no_procesadas):
        tiempo_actual = orden.placement_time
        print(f"Tiempo: {tiempo_actual.strftime('%H:%M:%S')}: La Plataforma ha recibido la orden {orden.order_id}")

        tienda = self.obtener_tienda_por_id(orden.store_id)
        if tienda is not None:
            print(f"Tiempo: {tiempo_actual.strftime('%H:%M:%S')}: La Plataforma ha informado a la tienda {orden.store_id}")
                         
        dist_tienda_cliente = great_circle((tienda['pick_up_lat'], tienda['pick_up_lng']),
                                           (orden.drop_off_lat, orden.drop_off_lng)).kilometers
        
        print(f"Distancia entre la tienda y el cliente: {dist_tienda_cliente:.2f}")
        capacidad_domiciliario = 5
                
        if lista_salida is None:
            lista_salida = []
        if lista_no_procesadas is None:
            lista_no_procesadas = []
            
        domiciliarios_disponibles = self.obtener_domiciliarios_disponibles(orden, capacidad_domiciliario)

        if domiciliarios_disponibles:
            domiciliario_mas_cercano = None
            distancia_mas_cercana = float('inf')

            for domiciliario in domiciliarios_disponibles:
                domiciliario_id = domiciliario['courier_id']
                dist = great_circle((orden.drop_off_lat, orden.drop_off_lng),
                                    (domiciliario['on_lat'], domiciliario['on_lng'])).kilometers

                if dist < distancia_mas_cercana:
                    domiciliario_mas_cercano = domiciliario
                    distancia_mas_cercana = dist

            if domiciliario_mas_cercano is not None:
                domiciliario_id = domiciliario_mas_cercano['courier_id']                
                
        #Definir el costo según la distancia
        costo_fijo_por_km = 1000 
        costo_por_distancia = dist_tienda_cliente * costo_fijo_por_km
        print(f"Costo según la distancia: {costo_por_distancia:.2f}")
        
        tercera_parte_costo_por_distancia = costo_por_distancia / 3
        
        incremento_por_rechazo = costo_por_distancia / 10 # Calcular la décima parte del costo por distancia
        
        num_rechazos = self.rechazos_por_orden.get(orden.order_id, 0) # Obtener el número de rechazos previos de esta orden
        
        # Ajustar el valor ofrecido basado de acuerdo al costo por distancia
        if tercera_parte_costo_por_distancia > 2000:
            valor_ofrecido_domicilio = tercera_parte_costo_por_distancia
        else:
            valor_ofrecido_domicilio = 2000
        
        print(f"Valor ofrecido por domicilio {valor_ofrecido_domicilio:.2f}")
                
        # Ajustar la probabilidad de rechazo basado en el costo por distancia

        # Crear un objeto de RechazoDomiciliario con la probabilidad ajustada
        rechazo_domiciliario = RechazoDomiciliario()

        for domiciliario in domiciliarios_disponibles:
            domiciliario_id = domiciliario['courier_id']
            
            # RECHAZOS -- > Inicializar contador de rechazos si es necesario
            if orden.order_id not in self.rechazos_por_orden:
                self.rechazos_por_orden[orden.order_id] = 0
                
            if valor_ofrecido_domicilio >= costo_por_distancia:
                probabilidad_rechazo = 0.2
            else:  
                probabilidad_rechazo = 0.8

            if random.random() <= probabilidad_rechazo:
                # Incrementar el contador de rechazos
                num_rechazos += 1
                self.rechazos_por_orden[orden.order_id] = num_rechazos
                valor_ofrecido_domicilio += incremento_por_rechazo * (self.rechazos_por_orden[orden.order_id])
                tiempo_actual += timedelta(minutes=1.5)
                print(f"Domiciliario {domiciliario_id} ha rechazado la orden {orden.order_id}. {num_rechazos}")
                print(f"Valor ofrecido por domicilio {valor_ofrecido_domicilio:.2f}")
                

                if self.rechazos_por_orden[orden.order_id] >= 4:
                    print(f"Orden {orden.order_id} rechazada definitivamente.")
                    self.ordenes_perdidas.append(orden.order_id)

                    costo_orden_perdida = random.uniform(40000, 100000)
                    ganancia_tienda_perdida = costo_orden_perdida
                    ganancia_plataforma_perdida = costo_orden_perdida * random.gauss(0.1, 0.2)
                    ganancia_domiciliario_perdida = valor_ofrecido_domicilio
                    costo_usuario_perdida = costo_orden_perdida + ganancia_domiciliario_perdida
                    
                    #Acumular ganancias perdidas por rechazos definitivos
                    self.ganancias_perdidas_tiendas += costo_orden_perdida
                    self.ganancias_perdidas_plataforma += ganancia_plataforma_perdida
                    self.ganancias_perdidas_domiciliarios += ganancia_domiciliario_perdida
                    break  # Salir del bucle, la orden ha sido rechazada definitivamente

            else:
                print(f"Tiempo: {tiempo_actual.strftime('%H:%M:%S')}: Domiciliario {domiciliario_id} asignado a la orden {orden.order_id}")
                
            #CUANDO ES ASIGNADO SE REALIZA TOOOODO LO DEMÁS PARA LLEVAR A CABO EL DOMICILIO                        
            # Asignar la velocidad del domiciliario según el tipo de vehículo
                tipo_vehiculo = domiciliario['vehicle']
                if tipo_vehiculo == 'motorcycle':
                    velocidad_domiciliario = 23
                elif tipo_vehiculo == 'bicycle':
                    velocidad_domiciliario = 12
                elif tipo_vehiculo == 'car': 
                    velocidad_domiciliario = 20
                elif tipo_vehiculo == 'walking':
                    velocidad_domiciliario = 5
                else:
                    velocidad_domiciliario = 30  # Velocidad por defecto si no se especifica el tipo de vehículo
                
                tiempo_ida_domiciliario = distancia_mas_cercana / velocidad_domiciliario
                tiempo_ida_domiciliario = timedelta(hours=tiempo_ida_domiciliario)
                print(f"Tiempo de ida del domiciliario a la tienda: {tiempo_ida_domiciliario}")
                
                time_difference = orden.ready_time - orden.placement_time
                
                tiempo_maximo = max(tiempo_ida_domiciliario, time_difference)
                
                tiempo_a_cliente = dist_tienda_cliente / velocidad_domiciliario
                tiempo_a_cliente = timedelta(hours=tiempo_a_cliente)
                print(f"Tiempo estimado del domiciliario al cliente: {tiempo_a_cliente} horas")

                tiempo_entrega = tiempo_ida_domiciliario + tiempo_a_cliente + tiempo_maximo
                tiempo_entrega = round(tiempo_entrega.total_seconds() / 60, 2)
                print(f"Tiempo estimado de entrega: {tiempo_entrega} minutos")
                tiempo_entrega_timedelta = timedelta(minutes=tiempo_entrega)
                
                tiempo_vuelta_base = tiempo_ida_domiciliario + tiempo_a_cliente   # Tiempo de regreso a la base
                print(f"Tiempo estimado en regreso al punto inicial el domiciliario: {tiempo_vuelta_base} horas")
                
                tiempo_disponibilidad_domiciliario = tiempo_entrega_timedelta + tiempo_actual + tiempo_vuelta_base
                print(f"Domiciliario {domiciliario_id} estará disponible a las {tiempo_disponibilidad_domiciliario.strftime('%H:%M:%S')}")
                
                costo_orden = random.uniform(40000, 100000)
                pago_tienda_a_plataforma = costo_orden * random.uniform(0.1, 0.2)
                ganancia_tienda = costo_orden - pago_tienda_a_plataforma
                costo_servicio = costo_orden * random.gauss(0.1, 0.03)
                ganancia_domiciliario = valor_ofrecido_domicilio
                costo_usuario = costo_orden + ganancia_domiciliario + costo_servicio
                if costo_por_distancia > ganancia_domiciliario:
                    sobrante_costo_por_distancia = costo_por_distancia - ganancia_domiciliario
                else:
                    sobrante_costo_por_distancia = 0
                ganancia_plataforma = pago_tienda_a_plataforma + sobrante_costo_por_distancia + costo_servicio
                                    
                # Actualizar las ganancias totales
                self.ganancia_total_domiciliarios += ganancia_domiciliario
                self.ganancia_total_tiendas += ganancia_tienda
                self.ganancia_total_plataforma += ganancia_plataforma
                self.ventas_totales += costo_usuario

                lista_salida.append({
                    'Orden': orden.order_id,
                    'Domiciliario': domiciliario_id,
                    'Tiempo de Ida': tiempo_ida_domiciliario.total_seconds() / 3600,
                    'Tiempo en Tienda': orden.preparation_time,
                    'Tiempo de Entrega': tiempo_entrega,
                    'Ganancia Domiciliario': ganancia_domiciliario,
                    'Costo orden': costo_orden,
                    'Costo de servicio':costo_servicio,
                    'Tienda': tienda,
                    'Ganancia tienda':ganancia_tienda,
                    'Pago tienda a plataforma':pago_tienda_a_plataforma,
                    'Costo por Distancia': costo_por_distancia
                })

                print(f"Ganancia del domiciliario: {ganancia_domiciliario:.2f}")
                print(f"Costo Total al Usuario: {costo_usuario:.2f}")
                print(f"Los otros costos son: costo de la orden: {costo_orden}, pago tienda a plataforma: {pago_tienda_a_plataforma:.2f}, Ganancia de la tienda: {ganancia_tienda:.2f}, Ganancia de la plataforma: {ganancia_plataforma:.2f}, sobrante costo por distancia: {sobrante_costo_por_distancia:.2f}")

                self.tiempo_disponibilidad[domiciliario_id] = tiempo_disponibilidad_domiciliario

                break
                   

#Cargar archivos CSV
carpeta_principal = 'Instancias'
subcarpeta = 'Instancia_3'

ruta_domiciliarios = os.path.join(carpeta_principal, subcarpeta, 'couriers.csv')
ruta_tiendas = os.path.join(carpeta_principal, subcarpeta, 'stores.csv')
ruta_ordenes = os.path.join(carpeta_principal, subcarpeta, 'orders.csv')

# Leer los archivos CSV
df_domiciliarios = pd.read_csv(ruta_domiciliarios)
df_tiendas = pd.read_csv(ruta_tiendas)
df_ordenes = pd.read_csv(ruta_ordenes)

# Define el tiempo de inicio y finalización de la simulación
tiempo_inicio_simulacion = datetime.strptime("00:00:00", "%H:%M:%S")
tiempo_finalizacion_simulación = datetime.strptime("23:59:00", "%H:%M:%S")

# Filtra las órdenes en el rango de tiempo
df_ordenes['placement_time'] = pd.to_datetime(df_ordenes['placement_time'])
df_ordenes['ready_time'] = pd.to_datetime(df_ordenes['ready_time'])
df_ordenes['expected_drop_off_time'] = pd.to_datetime(df_ordenes['expected_drop_off_time'])
df_ordenes_filtradas = df_ordenes[
    (df_ordenes['placement_time'].dt.strftime('%H:%M:%S') >= tiempo_inicio_simulacion.strftime('%H:%M:%S')) &
    (df_ordenes['placement_time'].dt.strftime('%H:%M:%S') <= tiempo_finalizacion_simulación.strftime('%H:%M:%S'))
]

# Crear un entorno de simulación
env = simpy.Environment()

# Crear listas para almacenar los resultados
lista_salida = []
lista_no_procesadas = []

# Simular el procesamiento de órdenes
platform_time = 2

plataforma = Plataforma(env, df_domiciliarios, df_tiendas, platform_time)

# Inicializar el tiempo actual al inicio de la simulación
tiempo_actual = tiempo_inicio_simulacion

# Definir la clase Orden
class Orden:
    def __init__(self, env, order_id, store_id, drop_off_lat, drop_off_lng, placement_time, preparation_time, ready_time, expected_drop_off_time):
        self.env = env
        self.order_id = order_id
        self.store_id = store_id
        self.drop_off_lat = drop_off_lat
        self.drop_off_lng = drop_off_lng
        self.placement_time = datetime.strptime(placement_time, "%H:%M:%S")
        self.preparation_time = preparation_time
        self.ready_time = datetime.strptime(ready_time, "%H:%M:%S")
        self.expected_drop_off_time = datetime.strptime(expected_drop_off_time, "%H:%M:%S")

    def tiempo_entrega_estimado(self):
        tiempo_ida_tienda = great_circle((df_tiendas.at[self.store_id, 'pick_up_lat'], df_tiendas.at[self.store_id, 'pick_up_lng']), (self.drop_off_lat, self.drop_off_lng)).kilometers / 30
        tiempo_ida_tienda = timedelta(hours=tiempo_ida_tienda)
        tiempo_entrega_estimado = tiempo_ida_tienda + timedelta(minutes=self.preparation_time)
        return tiempo_entrega_estimado

# Procesar todas las órdenes en df_ordenes_filtradas
for index, orden_row in df_ordenes_filtradas.iterrows():
    placement_time = orden_row['placement_time']
    ready_time = orden_row['ready_time']
    expected_drop_off_time = orden_row['expected_drop_off_time']

    orden = Orden(
        env,
        orden_row['order_id'],
        orden_row['store_id'],
        orden_row['drop_off_lat'],
        orden_row['drop_off_lng'],
        orden_row['placement_time'].strftime("%H:%M:%S"),  # Convierte a cadena
        random.randint(5, 15),
        orden_row['ready_time'].strftime("%H:%M:%S"),  # Convierte a cadena
        orden_row['expected_drop_off_time'].strftime("%H:%M:%S")  # Convierte a cadena
    )

    plataforma.procesar_orden(orden, lista_salida, lista_no_procesadas)

# Realizar la simulación sin restricciones de tiempo
env.run()

#Imprimir ganancias y ventas totales
print(f"Ganancia Total Domiciliarios: {round(plataforma.ganancia_total_domiciliarios, 2)}")
print(f"Ganancia Total Tiendas: {round(plataforma.ganancia_total_tiendas,2)}")
print(f"Ganancia Total Plataforma: {round(plataforma.ganancia_total_plataforma,2)}")
print(f"Ventas Totales (Costo Total al Usuario): {round(plataforma.ventas_totales,2)}")


# Imprimir el número de pedidos rechazados
print("Órdenes Perdidas:", plataforma.ordenes_perdidas)

#Imprimir ganancias perdidas por rechazos definitivos
print(f"Ganancias perdidas - Domiciliarios: {plataforma.ganancias_perdidas_domiciliarios}")
print(f"Ganancias perdidas - Tiendas: {plataforma.ganancias_perdidas_tiendas}")
print(f"Ganancias perdidas - Plataforma: {plataforma.ganancias_perdidas_plataforma}")

### EXPORTAR RESULTADOS A EXCEL

# ----> Para Domiciliarios

# Crear un DataFrame a partir de lista_salida
df = pd.DataFrame(lista_salida)

# Contar cuántas órdenes hizo cada domiciliario
orden_por_domiciliario = df['Domiciliario'].value_counts().reset_index()
orden_por_domiciliario.columns = ['Domiciliario', 'Cantidad de Órdenes']

ganancia_por_domiciliario = df.groupby('Domiciliario')['Ganancia Domiciliario'].sum().reset_index()
ganancia_por_domiciliario.columns = ['Domiciliario', 'Ganancia Total']

df_domiciliarios.set_index('courier_id', inplace=True)
df_final_domiciliarios = df_domiciliarios.join(orden_por_domiciliario.set_index('Domiciliario'))
df_final_domiciliarios = df_final_domiciliarios.join(ganancia_por_domiciliario.set_index('Domiciliario'))

# Reordenar y renombrar las columnas para el formato deseado
df_final_domiciliarios = df_final_domiciliarios[['vehicle', 'Cantidad de Órdenes', 'Ganancia Total']]
df_final_domiciliarios.reset_index(inplace=True)
df_final_domiciliarios.columns = ['courier_id', 'Tipo de Vehiculo', 'Total de Ordenes Atendidas', 'Ganancia Total']

# -----> Para ordenes

# Convertir lista_salida a DataFrame y realizar los cálculos necesarios
df_ordenes_completadas = pd.DataFrame(lista_salida)
df_ordenes_completadas['Hora de Entrega'] = df_ordenes_completadas['Tiempo de Ida'] + df_ordenes_completadas['Tiempo en Tienda'] + df_ordenes_completadas['Tiempo de Entrega']
df_ordenes_completadas['Hora de Entrega'] = df_ordenes_completadas['Hora de Entrega'].apply(lambda x: timedelta(hours=x) + tiempo_inicio_simulacion)

# Unir con df_ordenes para obtener los costos y rechazos
df_rechazos = pd.DataFrame.from_dict(plataforma.rechazos_por_orden, orient='index').reset_index()
df_rechazos.columns = ['Orden', 'Número de Rechazos']

# Unir df_rechazos con df_final_ordenes
df_final_ordenes = df_ordenes_completadas.merge(df_rechazos, on='Orden', how='left')

df_final_ordenes['Hora de Entrega'] = pd.to_datetime(df_final_ordenes['Hora de Entrega'])
df_final_ordenes['Hora de Entrega'] = df_final_ordenes['Hora de Entrega'].dt.strftime('%H:%M:%S')

# Seleccionar y renombrar columnas
df_final_ordenes = df_final_ordenes[['Orden', 'Costo orden', 'Número de Rechazos', 'Costo de servicio', 'Tiempo de Entrega', 'Hora de Entrega', 'Costo por Distancia', 'Ganancia Domiciliario']]
#df_final_ordenes.reset_index(inplace=True)
df_final_ordenes.columns = ['Order ID', 'Costo de la Orden', 'Número Total de Rechazos', 'Costo de servicio', 'Tiempo de Entrega', 'Hora de Entrega', 'Costo justo por distancia', 'Ganancia del domiciliario']


# --------> Para tiendas

df_ordenes_completadas['store_id'] = df_ordenes_completadas['Tienda'].apply(lambda x: x['store_id'])
pedidos_por_tienda = df_ordenes_completadas['store_id'].value_counts().reset_index()
pedidos_por_tienda.columns = ['store_id', 'Cantidad de Pedidos']

# Calcular la ganancia total y el pago a la plataforma por cada tienda
df_ganancias_tiendas = df_ordenes_completadas.groupby('store_id').agg({'Ganancia tienda':'sum', 'Pago tienda a plataforma':'sum'}).reset_index()

# Unir la información en un solo DataFrame
df_info_tiendas = pd.merge(pedidos_por_tienda, df_ganancias_tiendas, on='store_id')

# ---------> Para plataforma y total

# Crear un diccionario con los datos de la plataforma
datos_plataforma = {
    'Descripción': ['Ganancia plataforma', 'Ventas totales', 'Ganancia total domiciliarios', 'Ganancia total tiendas', 'Costo de oportunidad por ordenes rechazadas definitivamente', 'Total ordenes rechazadas definitivamente', 'Ganancias perdidas tiendas', 'Ganancias perdidas plataforma', 'Ganancias perdidas domiciliarios'],
    'Valor': [
        plataforma.ganancia_total_plataforma, 
        plataforma.ventas_totales, 
        plataforma.ganancia_total_domiciliarios, 
        plataforma.ganancia_total_tiendas, 
        '', # Costo de oportunidad no se calculó en el código, así que se deja en blanco o se agrega un cálculo si lo tienes
        len(plataforma.ordenes_perdidas), 
        plataforma.ganancias_perdidas_tiendas, 
        plataforma.ganancias_perdidas_plataforma, 
        plataforma.ganancias_perdidas_domiciliarios
    ]
}

# Convertir el diccionario en un DataFrame
df_plataforma = pd.DataFrame(datos_plataforma)

#---------> Para gráficos

#Domiciliarios por Hora
# Convertir las columnas 'on_time' y 'off_time' a tipo datetime
df_domiciliarios['on_time'] = pd.to_datetime(df_domiciliarios['on_time'], format='%H:%M:%S').dt.time
df_domiciliarios['off_time'] = pd.to_datetime(df_domiciliarios['off_time'], format='%H:%M:%S').dt.time

# Crear un DataFrame para cada hora del día
hours = [hour.time() for hour in pd.date_range("00:00", "23:59", freq="H")]
availability_df = pd.DataFrame(hours, columns=['hour'])

# Contar los domiciliarios disponibles para cada hora
availability_df['count'] = availability_df['hour'].apply(
    lambda hour: ((df_domiciliarios['on_time'] <= hour) & (df_domiciliarios['off_time'] >= hour)).sum()
)

# Crear un gráfico para mostrar la disponibilidad de domiciliarios por hora
plt.figure(figsize=(12, 6))
plt.bar(availability_df['hour'].astype(str), availability_df['count'])
plt.title('Disponibilidad de Domiciliarios por Hora')
plt.xlabel('Hora')
plt.ylabel('Cantidad de Domiciliarios Disponibles')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.tight_layout()

## Ordenes por hora
# Convertir la columna 'placement_time' a tipo datetime
df_ordenes['placement_time'] = pd.to_datetime(df_ordenes['placement_time'])

#Ajustar el rango de tiempo para incluir todas las horas del día de 00:00 a 23:00
full_hours = pd.date_range(start='00:00', end='23:00', freq='H')
full_hours_str = [hour.strftime('%H:%M') for hour in full_hours]
# Agrupar por hora y contar las ordenes
order_counts_per_hour = df_ordenes['placement_time'].dt.floor('H').value_counts().sort_index()

# Asegurar que todas las horas del día están representadas en el gráfico de órdenes
order_counts_per_hour_full = order_counts_per_hour.reindex(full_hours, fill_value=0)

# Creando el gráfico
plt.figure(figsize=(12, 6))
order_counts_per_hour.plot(kind='bar')
plt.xticks(ticks=range(len(order_counts_per_hour)), labels=[t.strftime('%H:%M') for t in order_counts_per_hour.index])
plt.title('Cantidad de Órdenes por Hora')
plt.xlabel('Hora')
plt.ylabel('Cantidad de Órdenes')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()

rechazos_por_orden = plataforma.rechazos_por_orden
ordenes_perdidas = plataforma.ordenes_perdidas

# Categorizar las órdenes
ordenes_aceptadas_sin_rechazo = sum(1 for rechazos in rechazos_por_orden.values() if rechazos == 0)
ordenes_rechazadas_una_vez = sum(1 for rechazos in rechazos_por_orden.values() if rechazos == 1)
ordenes_rechazadas_dos_veces = sum(1 for rechazos in rechazos_por_orden.values() if rechazos == 2)
ordenes_rechazadas_tres_veces = sum(1 for rechazos in rechazos_por_orden.values() if rechazos == 3)
ordenes_rechazadas_definitivamente = len(ordenes_perdidas)

# Datos de la tabla
datos_tabla = {
    "Categoría": ['Sin Rechazo', '1 Rechazo', '2 Rechazos','3 Rechazos', 'Rechazadas Definitivamente'],
    "Cantidad de Órdenes": [ordenes_aceptadas_sin_rechazo, ordenes_rechazadas_una_vez, ordenes_rechazadas_dos_veces, 
                            ordenes_rechazadas_tres_veces, ordenes_rechazadas_definitivamente]
}

# Crear DataFrame
df = pd.DataFrame(datos_tabla)

# Exportar gráfico a una imagen
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(df["Categoría"], df["Cantidad de Órdenes"], color=['blue', 'green', 'orange', 'red'])
plt.title('Distribución de Órdenes Basada en Rechazos')
plt.xlabel('Categoría')
plt.ylabel('Cantidad de Órdenes')


output_folder = 'outputs_Bin'
excel_path = os.path.join(output_folder, 'resultadosIns_3.xlsx')

# Guardar gráfico como imagen
ruta_imagen_1 = os.path.join(output_folder, 'grafico_rechazosIns_3.png')

plt.savefig(ruta_imagen_1)


# Añadir el grafico de cantidad de ordenes y domiciliarios
plt.figure(figsize=(15, 8))
plt.plot(full_hours_str, order_counts_per_hour_full.values, label='Órdenes por Hora', color='blue', marker='o', linestyle='-', linewidth=2)
plt.plot(full_hours_str, availability_df['count'], color='red', marker='o', linestyle='-', linewidth=2, label='Domiciliarios Disponibles')
plt.title('Órdenes y Domiciliarios Disponibles por Hora')
plt.xlabel('Hora')
plt.ylabel('Cantidad')
plt.xticks(full_hours_str, rotation=90)
plt.grid(axis='y')
plt.legend()
plt.tight_layout()

# Guardar el gráfico como imagen
ruta_imagen_2 = os.path.join(output_folder, 'grafico_DisponiblesIns_3.png')
plt.savefig(ruta_imagen_2)


with pd.ExcelWriter(excel_path, engine='openpyxl') as writer:
    df_final_domiciliarios.to_excel(writer, sheet_name='Domiciliarios', index=False)
    df_final_ordenes.to_excel(writer, sheet_name='Órdenes', index=False)
    df_info_tiendas.to_excel(writer, sheet_name='Tiendas', index=False)
    df_plataforma.to_excel(writer, sheet_name='Plataforma', index=False)
    df.to_excel(writer, sheet_name='Rechazos', index=False)
        # Añadir el gráfico de rechazos
book = openpyxl.load_workbook(excel_path)
sheet = book['Rechazos']

# Crea un objeto Image con la ruta de la imagen
img1 = Image(ruta_imagen_1)

# Inserta la imagen en la celda deseada
sheet.add_image(img1, 'G1')
img1.width = img1.width * 0.5  
img1.height = img1.height * 0.5


# Cargar e insertar la segunda imagen
img2 = Image(ruta_imagen_2)
sheet.add_image(img2, 'Q1')  
img2.width = img2.width * 0.5  
img2.height = img2.height * 0.5 

# Guarda los cambios en el archivo
book.save(excel_path)

rechazos_por_orden = plataforma.rechazos_por_orden
ordenes_perdidas = plataforma.ordenes_perdidas

# Mostrar la tabla de rechazos
print(f"ordenes_aceptadas_sin_rechazo: {ordenes_aceptadas_sin_rechazo}")
print(f"ordenes_rechazadas_una_vez: {ordenes_rechazadas_una_vez}")
print(f'ordenes_rechazadas_dos_veces: {ordenes_rechazadas_dos_veces}')
print(f'ordenes_rechazadas_tres_veces: {ordenes_rechazadas_tres_veces}')
print(f'ordenes_rechazadas_definitivamente: {ordenes_rechazadas_definitivamente}')
```
